```python
import requests, base64, os, threading, time, queue
#API
tpath = 'http://api.95man.com:8888/api/Http/UserTaken?user=Snion&pwd=GB5994413&isref=0'
taken = requests.get(tpath).text[2:]
api = "http://api.95man.com:8888/api/Http/Recog?Taken="+taken+"&imgtype=1&len=4"

# 下载验证码
def getcode(path):
    vpath = "https://yz.scu.edu.cn/User/Valicdoe?t=6666"
    try:
        img = requests.get(vpath)
    except:
        print("验证码爬取错误")
        exit()

    with open(path, 'wb') as f:
        try:f.write(img.content)
        except:
            print("文件写入错误")
            exit()

def verifycode(path):
    #API识别
    with open(path, 'rb') as f:
        img = base64.b64encode(f.read())
    datas = {"ImgBase64":img}
    code = requests.post(api, data=datas).text[8:12]
    code = code.replace("O","0")
    try:
        os.rename(path, path[:-5]+"_"+code+".png")
    except:pass

class myThread (threading.Thread):
    def __init__(self, ID, num):
        threading.Thread.__init__(self)
        self.name = ID
        self.num= num
    def run(self):
        k = self.num
        path = './sample/origin/'+str(k)+'.tiff'
        getcode(path)
        verifycode(path)
        print ("Thread end："+str(self.name) +"---"+ str(k))
        


def main():
    n=333 # 基数
    for i in range(0,100,4):
        threads = []
        for j in range(4):
            th = myThread(j, i+n+j)
            th.start()
            threads.append(th)
            time.sleep(0.3)
        for th in threads:
            th.join(2)  ##超时时间

    
main()
```
